```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv.zip')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group').count()

#5 neighborhood groups available. Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(["neighbourhood_group", "room_type"]).count()

#No. entire home/apartment is most popular
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb[["host_id", "id", "number_of_reviews"]].groupby("host_id").count().sort_values("number_of_reviews", ascending = False)

air_bnb.loc[air_bnb["host_id"] == 219517861]

#Sonder(NYC)
```

```python
#Which neighorhood group has the highest average price?
air_bnb[["neighbourhood_group", "price"]].groupby("neighbourhood_group").mean()

#Manhattan
```

```python
# Which neighbor hood group has the highest total price?
air_bnb[["neighbourhood_group", "price"]].groupby("neighbourhood_group").sum()

#Manhattan
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_id').agg({'id': 'count', 'price': "sum"}).sort_values("id", ascending = False).head()

air_bnb.loc[air_bnb["host_id"] == 219517861]

#Sonder (NYC)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

avail_review = air_bnb.loc[(air_bnb["availability_365"] == 0) & (air_bnb["number_of_reviews"] >= 100)]
no_avail_many_reviews = avail_review["host_id"].unique()
no_avail_many_reviews
```

```python
# What host has the highest total of prices and where are they located?

air_bnb[["host_id", "price"]].groupby("host_id").sum().sort_values("price", ascending = False).head()

#air_bnb.loc[air_bnb["host_id"] == 30283594]

#1. Sonders (NYC)
#2. Blueground
#3. Sally
#4. Red Awning
#5. Kara
```

```python
# When did Danielle from Queens last receive a review?

air_bnb.loc[(air_bnb["host_name"] == "Danielle") & (air_bnb["neighbourhood_group"] == "Queens")]

#Danielle host id 26432133: 7/8/2019
#Danielle host id 18051286: No reviews
#Danielle host id 154256662: 1/2/2018
#Danielle host id 201647469: 6/20/2019
```
